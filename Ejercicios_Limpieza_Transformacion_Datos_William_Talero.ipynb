## Ejercicio Fechas Publicación

Teniendo en cuenta que existe una Fecha de Publicación del Proceso y una fecha de última publicación, realice la limpieza de formato de dichas variables. Defina una nueva variable en la que se calcule la diferencia de éstas dos fechas y responda si incluiría las dos fechas dentro de la fuente de datos.

*Utilice estadísticas descriptivas para responder la pregunta*

#**Solucion** 

Para este punto cargamos la base que se exporto al final de la clase 6 de limpieza de datos en la cual se realizo lo siguiente:

**Completitut**: Se borraron los registros perdidos, y se imputaron los datos faltantes.

**Coherencia**: Se eliminaron los registros duplicados.

**Consistencia**: Se verifico que los que en la variable de Nit presente una longitud de 9 digitos.


```python
df_SECOP2 = pd.read_csv('/content/drive/MyDrive/Datos_SECOO_LT.csv', 
                 sep = ',',
                 thousands=',', decimal='.',
                 skip_blank_lines=True,
                 low_memory = False,
                 dtype={
                     'Nit Entidad': str
                 })
df_SECOP2.head()
```

```python
df_SECOP2['Fecha de Publicacion del Proceso'] = pd.to_datetime(df_SECOP2['Fecha de Publicacion del Proceso'],format='%Y/%m/%d')
df_SECOP2['Fecha de Publicacion del Proceso'].head()
```

```python
df_SECOP2['Fecha de Ultima Publicación'] = pd.to_datetime(df_SECOP2['Fecha de Ultima Publicación'],format='%Y/%m/%d')
df_SECOP2['Fecha de Ultima Publicación'].head()
```

```python
df_SECOP2["Diferencia"]=(df_SECOP2["Fecha de Publicacion del Proceso"] - df_SECOP2["Fecha de Ultima Publicación"]).dt.days
df_SECOP2['Diferencia'].head()
```

```python
df_SECOP2['Diferencia'].describe()
```

```python
(df_SECOP2 
  .groupby("Diferencia")
  .agg(frequency=("Diferencia", "count")))
```

***Respuesta***: Decididria no incluir las 2 variables ya que la diferencia entre las 2 fechas para la mayoria de registros no supera 1 dia, por lo que podriamos dejar solo una variable.


## Ejercicio Tipo de Contrato y Subtipo de Contrato
De acuerdo con las variables Contrato y Subcontrato, realice una tabla cruzada y responda ¿Vale la pena dejar ambas variables en la fuente de datos?

#Solucion

```python
pd.crosstab(index=df_SECOP2['Tipo de Contrato'],
            columns=df_SECOP2['Subtipo de Contrato'], margins=True)
```

```python
(df_SECOP2 
  .groupby("Subtipo de Contrato")
  .agg(frequency=("Subtipo de Contrato", "count")))
```

***Respuesta***: No vale la pena ya que en la variable **Subtipo de Contrato** solo tiene un unico nivel. 

## Ejercicio Duración del contrato
Teniendo en cuenta que la duración de los contratos tiene métricas diferentes, encuentre una estrategia para crear una nueva variable en la cuál se encuentre parametrizada la variable de duración del contrato.

```python
df_SECOP2['Unidad de Duracion'].value_counts()
```

```python
df_SECOP2[['Unidad de Duracion','Duracion']].sample(20)
```

### **Solucion**


```python
(df_SECOP2 
  .groupby("Unidad de Duracion")
  .agg(frequency=("Unidad de Duracion", "count")))
```

```python
len(df_SECOP2["Unidad de Duracion"])
```

```python
c=df_SECOP2['Unidad de Duracion']
a = np.zeros(shape=850953)
b = df_SECOP2["Duracion"]
```

```python
for i in range(len(c)):
  if c[i] == "Años":
    a[i]=round(b[i]*12,2)
  elif c[i] == "Meses":
    a[i]=round(b[i],2)
  elif c[i] == "Dias": 
    a[i]=round(b[i]/30,2)
  elif c[i] == "ND":
    a[i]=0
    
```

```python
df_SECOP2['Meses']=a;df_SECOP2["Meses"].head
```
