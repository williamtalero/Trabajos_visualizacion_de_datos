# Visualizacion 1

Los datos provienen de una muestra representativa de 920 adolescentes encuestados entre septiembre y noviembre de 2018. a cada adolescente se le relizaron prguntas de tipo dicotomicas que consistia en si cada problematica se presenta en mayor o menor medida entre personas de su edad que pertenezcan a su comunidad.

visualizacion_salud_mental.png

*Este gráfico, apareció en el New York Times en el artículo del 20 de febrero de 2019 " El problema principal de la salud mental " es del informe del Centro de investigación Pew " La mayoría de los adolescentes estadounidenses ven la ansiedad y la depresión como un problema importante entre sus compañeros".*

#**¿Que? Abstraccion de datos**

**¿Que datos ve el usuario?** El usuario observa como los adolescentes consideran que ciertos problemas se presentan en mayor o menor medida entre las personas de su edad en su comunidad.

**¿Que datos se proporcionan** A partir de la visualizacion de barras apiladas observamos las 7 problematicas, y cada problematica con sus correspondientes 3 barras que nos permiten saber el porcentaje de adolecentes que perciben este problema en mayor o menor medida.

**¿Que informacion se puede concluir a partir de los datos?** A partir de los datos se puede concluir que el problema que se presenta en mayor medida entre los adolescentes es la ansiedad y la depresion, por otro lado entre los adolescentes no se percibe en mayor cantidad problemas como la pobreza, pandillas y el embarazo en adolescentes.

#**¿por qué? ¿Que tipo de tarea requiere el usuario?**

A partir de los resultados de la encuesta se pretendia obtener variables dummy por cada variable, por lo que la tarea que requiere el usuario podriamos decir que es **identificar tendencias** ya que el objetivo final es evidenciar si existe una percepcion mayor o menor en los adolescentes para cada problematica.

#**¿Como? ¿que tipo de codificiacion visual se utilizo?**
El modismo que se utilizo para la visualizacion de datos es el de barras apiladas al 100%, en el que cada una de las 7 barras representan los problemas, a si mismo cada barra se encuetra separadas en segmentos de colores que en el que cada color representan la contribucion relativas de cada nivel de la variable categorica a la barra total.

#**Marcas y canales utilizados**

**Marcas :** Lineas.

**Canales :** *posicion* vertical, *colores* contrasre claro-oscuro en tono verde

#**Problemas y mejoras posibles**

##**Problemas**

**1**Al encontrar esta visualizacion la primera impresion que me da, es que es una visualizacion sin terminar en la primera barra apilada encuentro el signo de porcentaje que no se presenta en las demas barras ni en los demas segmentos de la misma, por lo que este dato al menos para las demas barras se podria confundir con la cantidad de adolescentes en cada nivel de la variable en vez del porcentaje en cada nivel de la variable

**2** Otro problema que encuentro es que a pesar de que las barras se encuentran alineadas parecen no tener el mismo tamaño, barras como la ansiedad y la depresion parecen ser mas largas que la adiccion a las drogas.

**3** Finalmente no considero que la paleta de colores sea la adecuada a la hora de visualizar los datos, ya que los datos son antagonicos, por lo que directamente me decantaria a usar 2 colores completamente opuestos. (Rojo y Azul).

##**Mejoras**

Cambiaria el modismo y presentaria 7 graficos de torta, ademas de expresar los valores con su respectivo porcentaje agregaria el valor lo que me permite saber que valor representa ese porcentaje, y cambiara la paleta de colores por una que me permita evidenciar que los datos son antagonicos. y separaria una rebanada que corresponde al valor que nos importa que es la cantidad de adolescentes que concideran que esta problematica se presenta en mayor medida

##*Ejemplo*

```python
import matplotlib.pyplot as plt


ansiedad_y_depresion = [644,239,37]
nombres = ["Major","Minor","No"]
desfase = (0.1, 0, 0)
plt.pie(ansiedad_y_depresion, labels=nombres, autopct="%0.1f %%", explode=desfase)
plt.axis("equal")
plt.show()
```

#Visualizacion 2

Aquí vemos la migración de 6 continentes: inmigración, emigración y migración interna a lo largo del año 2017. las sifras se encuentran en millones de personas.

visualizacion_enfermedades.png

*Este gráfico de migración de 2017 proviene del artículo del New York Times " Los migrantes están en aumento en todo el mundo, y los mitos sobre ellos están dando forma a las actitudes".*

#**¿Que? Abstraccion de datos**

**¿Que datos ve el usuario?** El usuario observa los flujos de inmigracion durante el año 2017.

**¿Que datos se proporcionan?** los datos se proporcionan del Departamento de Asuntos Económicos y Sociales de las Naciones Unidas.

**¿Que informacion se puede concluir a partir de los datos?** A partir de los datos se puede concluir que Asia es el continente que presenta mayor flujo de inmigrantes dentro del mismo por otro lado la mayor parte del total de los inmigrantes de norte america provienen de latino america.

#**¿por qué? ¿Que tipo de tarea requiere el usuario?**

Los datos que se muestran en la visualizacion corresponden a la cantidad de inmigrantes en millones de personas, la tarea que se requiere es **comparar tendencias** esto ya que se cuenta con la informacion de la cantidad de inmigrantes por continente incluyendo su continente de origen. 

#**¿Como? ¿que tipo de codificiacion visual se utilizo?**
El modismo que se utilizo para la visualizacion de flujos de inmigracion del año 2017 fue un diagrama de cuerdas esto para comparar los flujos de inmigracion de los 6 continentes presentes.

#**Marcas y canales utilizados**

**Marcas :** Lineas.

**Canales :** *size* longitud, *shape* continentes

#**Problemas y mejoras posibles**

##**Problemas**

**1** la unidad que se usa para comparacion es millones de inmigrantes, cuando el porcentaje de la cantidad total de inmigrantes  se podria utilizar de una mejor manera a la hora de realizar comparaciones.

**2** considero que a pesar de que el diagrama de cuerdas es ideal a la hora de realizar comparaciones cuando se comparte algo en comun en este caso los imigrantes, considero que existen otros modismos que pueden cumplir con la misma funcion de manera que el usuario logre captar toda la informacion que se le esta mostrando

##**Mejoras**

**1** Para las mejoras buscaria aplicar otro modismo realizar un grafico de barras ya sea un grafico de barras multiple, o un grafico de barras apiladas de manera que me permita comparar los flujos de inmigracion entre continentes, de igual manera asi se podria comparar caracteristicas de la poblacion migrante dentro de cada continente en este caso, su continente de origen.

**2** otra idea que se prodia plasmar seria ubicar 6 diagramas de anillos y ir variando el tamaño de cada anillo segun sea el total de flujo de migracion por continente

#*Ejemplo*

```python
import pandas as pd
import matplotlib.pyplot as plt

continente=["Latino America","Norte America","Europa","Asia","Oceania","Africa"]
data={
    "Latino America":[6.1,26.4,0,0,0,0],
    "Norte America":[1.4,1.2,1,0,0,0],
    "Europa":[1.3,7.6,41,7.1,3.1,1],
    "Asia":[0,17.2,20.5,63.3,3.2,0],
    "Oceania":[0,0,0,0,1.1,0],
    "Africa":[0,0,1,0,0,19.4]
}

df=pd.DataFrame(data,index=continente)

df.plot(kind="bar",stacked=True,figsize=(10,8))
plt.legend(loc="lower left",bbox_to_anchor=(0.8,1.0))
plt.show()
```
