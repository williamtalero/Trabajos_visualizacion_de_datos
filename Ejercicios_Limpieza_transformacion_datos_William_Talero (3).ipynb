## Ejercicio Duración del contrato
Teniendo en cuenta que la duración de los contratos tiene métricas diferentes, encuentre una estrategia para crear una nueva variable en la cuál se encuentre parametrizada la variable de duración del contrato.

```python
df_SECOP2['Unidad de Duracion'].value_counts()

#Dias     424960
#Meses    390226
#ND        29277
#Años       6490
#Name: Unidad de Duracion, dtype: int64
```

```python
df_SECOP2[['Unidad de Duracion','Duracion']].sample(20)
```

### **Solucion**


```python
(df_SECOP2 
  .groupby("Unidad de Duracion")
  .agg(frequency=("Unidad de Duracion", "count")))
```

```python
len(df_SECOP2["Unidad de Duracion"])
```

```python
c=df_SECOP2['Unidad de Duracion']
a = np.zeros(shape=850953)
b = df_SECOP2["Duracion"]
```

Realizamos un for que reccorra  df_SECOP2["Duracion"] de tal manera que la Duracion quede parametrizada en meses

```python
for i in range(len(c)):
  if c[i] == "Años":
    a[i]=round(b[i]*12,2)
  elif c[i] == "Meses":
    a[i]=round(b[i],2)
  elif c[i] == "Dias": 
    a[i]=round(b[i]/30,2)
  elif c[i] == "ND":
    a[i]=0
    
```

```python
df_SECOP2['Duracion']=a;df_SECOP2["Duracion"]
```

```python
df_SECOP2['Unidad de Duracion']=df_SECOP2['Unidad de Duracion'].replace(["Años","Dias"],["Meses","Meses"])
```

```python
(df_SECOP2 
  .groupby("Unidad de Duracion")
  .agg(frequency=("Unidad de Duracion", "count")))
```

Eliminamos los datos que tengan como unidad de duracion **ND**

```python
df_SECOP3=df_SECOP2[df_SECOP2["Unidad de Duracion"]=="Meses"]
```

```python
df_SECOP3["Duracion"].describe()
```

```python
print(df_SECOP3[['Unidad de Duracion','Duracion']])

#       Unidad de Duracion  Duracion
#0                   Meses      1.00
#1                   Meses     11.47
#2                   Meses      7.00
#4                   Meses      5.00
#5                   Meses      2.00
#...                   ...       ...
#850948              Meses      6.77
#850949              Meses      9.53
#850950              Meses      6.00
#850951              Meses     11.20
#850952              Meses      2.00
```

Como podemos ver ya se encuentra parametrizada la duracion en meses.
